```python
import cv2
import matplotlib.pyplot as plt

# --- 1. Ler a imagem ---
# Troque "lena.png" pelo caminho da sua imagem
img_colorida = cv2.imread("lena.png")

# OpenCV lê em BGR, convertemos pra RGB pra exibir certo no matplotlib
img_rgb = cv2.cvtColor(img_colorida, cv2.COLOR_BGR2RGB)

# --- 2. Converter para escala de cinza ---
img_cinza = cv2.cvtColor(img_colorida, cv2.COLOR_BGR2GRAY)

# --- 3. Binarizar (threshold) ---
# threshold = 127 significa que valores acima ficam brancos (255), abaixo ficam pretos (0)
_, img_binaria = cv2.threshold(img_cinza, 127, 255, cv2.THRESH_BINARY)

# --- 4. Exibir os resultados ---
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Imagem Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(img_cinza, cmap="gray")
plt.title("Imagem em Cinza")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(img_binaria, cmap="gray")
plt.title("Imagem Binária")
plt.axis("off")

plt.show()
```
